```python
import matplotlib.pyplot as plt
import numpy as np

starting_balance = 100
leverage = 0.1
loss = 0.4
gain = 0.5
num_games = 1000

def coinFlip(size, balance, leverage):
    xmin, xmax, dx = 1, size, 1
    x = np.arange(xmin, xmax, dx)
    balances = []
    
    for i in range(size):
        balances.append(balance)
        
        # Check if we can play another game
        if (balance <= 0): 
            break
            
        investment = leverage * balance
        balance = balance - investment
            
        # Play the game
        flip = np.random.randint(0, 2, size=1)
        if flip: investment = investment - loss * investment
        else: investment = investment + gain * investment
    
        balance = balance + investment
    
    #flips = np.random.randint(0, 2, size=size)
    return x, np.array(balances[:size-1])#[balances[:i].mean() for i in x]

x, y = coinFlip(num_games, starting_balance, leverage)
plt.plot(x, y)
```
